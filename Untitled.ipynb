```python
import pandas as pd
import numpy as np
import shap
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score
from xgboost import XGBClassifier
df = pd.read_csv(r"C:\Users\SNEHALATA ROUT\Downloads\collections.postmortem_tissue_core.rnaseq_metadata.csv")
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
)
df["age_at_symptom_onset"] = pd.to_numeric(
    df["age_at_symptom_onset"],
    errors="coerce"   # invalid strings → NaN
)

# target
df["early_onset"] = (df["age_at_symptom_onset"] <= 50).astype(int)
df = df.dropna(subset=["age_at_symptom_onset"])
demographic_features = [
    "sex",
    "ethnicity",
    "pct_african",
    "pct_european",
    "pct_south_asian",
    "pct_east_asian",
    "pct_americas"
]
genetic_features = [
    "c_9_orf_72_repeat_expansion",
    "atxn_2_repeat_expansion",
    "c_9_repeat_size",
    "atxn_2_repeat_size",
    "family_history_of_als_ftd",
    "reported_genomic_mutations"
]
clinical_features = [
    "site_of_motor_onset",
    "site_of_motor_onset_detail",
    "sex_genotype"
]
multimodal_features = (
    demographic_features +
    genetic_features +
    clinical_features
)
def train_evaluate_model(df, feature_list, model_name):
    X = df[feature_list]
    y = df["early_onset"]

    # one-hot encoding for categorical variables
    X = pd.get_dummies(X, drop_first=True)

    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size=0.2,
        stratify=y,
        random_state=42
    )

    model = XGBClassifier(
        n_estimators=300,
        max_depth=4,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        eval_metric="auc",
        random_state=42
    )

    model.fit(X_train, y_train)

    y_prob = model.predict_proba(X_test)[:, 1]
    y_pred = (y_prob >= 0.5).astype(int)

    auc = roc_auc_score(y_test, y_prob)
    acc = accuracy_score(y_test, y_pred)
    print(f"{model_name}")
    print(f"  AUC      : {auc:.3f}")
    print(f"  Accuracy : {acc:.3f}")
    print("-" * 40)
    return auc, acc, 

results = {}
results["Demographic"] = train_evaluate_model(
    df, demographic_features, "Unimodal – Demographic"
)

results["Genetic"] = train_evaluate_model(
    df, genetic_features, "Unimodal – Genetic"
)

results["Clinical"] = train_evaluate_model(
    df, clinical_features, "Unimodal – Clinical"
)

results["Multimodal"] = train_evaluate_model(
    df, multimodal_features, "Multimodal (D + G + C)"
)
results_df = pd.DataFrame.from_dict(
    results,
    orient="index",
    columns=["AUC", "Accuracy"]
)

print(results_df)



```

```python
import shap
import matplotlib.pyplot as plt
def train_evaluate_shap(df, feature_list, model_name):
    X = df[feature_list]
    y = df["early_onset"]

    # one-hot encode categorical features
    X = pd.get_dummies(X, drop_first=True)

    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size=0.2,
        stratify=y,
        random_state=42
    )

    model = XGBClassifier(
        n_estimators=300,
        max_depth=4,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        eval_metric="auc",
        random_state=42
    )

    model.fit(X_train, y_train)

    # predictions
    y_prob = model.predict_proba(X_test)[:, 1]

    auc = roc_auc_score(y_test, y_prob)
    acc = accuracy_score(y_test, (y_prob >= 0.5).astype(int))
    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(X_train)

    return model, X_train, shap_values, auc, acc


    print(f"{model_name} | AUC={auc:.3f}, Accuracy={acc:.3f}")

    
```

```python
models = {}
models["Demographic"] = train_evaluate_shap(
    df, demographic_features, "Unimodal – Demographic"
)

models["Genetic"] = train_evaluate_shap(
    df, genetic_features, "Unimodal – Genetic"
)

models["Clinical"] = train_evaluate_shap(
    df, clinical_features, "Unimodal – Clinical"
)

models["Multimodal"] = train_evaluate_shap(
    df, multimodal_features, "Multimodal (D + G + C)"
)
```
